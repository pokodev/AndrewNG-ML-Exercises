```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.io import loadmat
```

```python
X = loadmat('data/ex8data1.mat')['X']
```

```python
plt.figure(figsize=(15,10))
plt.scatter(X[:,0],X[:,1],marker='x',color='r',s=15)
```

```python
def estimate_gaussian(X):
    mu = X.mean(axis=0)
    sigma = X.var(axis=0)
    
    return mu, sigma
```

```python
mu, sigma = estimate_gaussian(X)
mu, sigma
```
