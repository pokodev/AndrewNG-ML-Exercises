```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

z=np.linspace(-5,5)
plt.plot(z,sigmoid(z))
```

```python
data.insert(0,'ones',1)
X=data.iloc[:,:3].values
y=data['admitted'].values
X.shape
theta = np.zeros(X.shape[1])
```

```python
#定义cost function
def costfunction(X,y,theta):
    m=X.shape[0]
    z=X.dot(theta)
    hypo = sigmoid(z)
    return sum(-y * np.log(hypo) - (1-y)*np.log(1-hypo))/m
```

```python
data.head()
```

```python
costfunction(X,y,theta)
```

```python
def gradient(X,y,theta):
    z=X.dot(theta)
    m=X.shape[0]
    error = sigmoid(z) - y
    return error.T.dot(X)/m
```

```python
def gradient_desent(X,y,theta,alpha,itera):
    _theta = theta.copy()
    cost_data = [costfunction(X,y,theta)]
    for i in range(itera):
        _theta=_theta - (alpha * gradient(X,y,_theta))
        cost_data.append(costfunction(X,y,_theta))
    return _theta,cost_data
```

```python
costfunction(X,y,theta)
```

```python
def cost(theta, X, y):
    ''' cost fn is -l(theta) for you to minimize'''
    return np.mean(-y * np.log(sigmoid(X @ theta)) - (1 - y) * np.log(1 - sigmoid(X @ theta)))

# X @ theta与X.dot(theta)等价
def gradient(theta, X, y):
#     '''just 1 batch gradient'''
    return (1 / len(X)) * X.T @ (sigmoid(X @ theta) - y)
```

```python
import scipy.optimize as opt
res = opt.minimize(fun=cost, x0=theta, args=(X, y), method='Newton-CG', jac=gradient)
res
```

```python
theta = res.x
theta
```

```python
def predict(theta, X):
    z = X.dot(theta)
    probability = sigmoid(z)
    return [1 if x >= 0.5 else 0 for x in probability]
pre=pd.DataFrame({'predict':predict(theta,X)})
data_pre=pd.concat([data,pre],1)
data_pre.head()
```

```python
#预测准确率
len(data[data_pre['admitted'] == data_pre['predict']]) / len(data) * 100
#very cool
```
